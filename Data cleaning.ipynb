### Script for automating it for all datasets

```python
datasets = ['demographic_modified.xlsx', 'diet_modified.xlsx', 'examination_modified.xlsx', 'labs_modified.xlsx', 'questionnaire_modified.xlsx']
datasets_check = ['demographic_modified.xlsx']
```

```python
for dataset in datasets:
    df = pd.read_excel(dataset)
    df = df.fillna('.')
    export_name = "cleaned_" + dataset[:-5]
    if(dataset == 'diet_modified.xlsx' or dataset == 'examination_modified.xlsx'):
        df = df.drop(columns=['SHORTENED NAME'])
        df = df.iloc[2:]
    elif(dataset == 'questionnaire_modified.xlsx'):
        df = df.iloc[1:]
    else:
        df = df.iloc[2:]
    #Reset index so you can .iloc easier
    df.reset_index(drop=True).head()
    #Set first row as column headers
    df = df.rename(columns=df.iloc[0]).drop(df.index[0]).reset_index(drop=True)
    #Check if there are periods in column name
    col_list = []
    for col in df.columns:
        try:
            col.index('.')
        except ValueError:
            col_list.append(col)
        else:
            col = col.replace('.', '_')
            col_list.append(col)
    df.columns = col_list
    #Append the column name to each value
    for col in df.columns[1:]:
        df.loc[:,col] = '{}_'.format(col) + df[col].astype(str)
    #Export the file to csv
    df.to_csv(export_name + ".csv", index=False)
    print(export_name)
    
```

### Script for automating the mapping to all cleaned datasets

```python
cleaned_datasets = ['cleaned_demographic_modified.csv', 'cleaned_diet_modified.csv', 'cleaned_examination_modified.csv', 'cleaned_labs_modified.csv', 'cleaned_questionnaire_modified.csv']
mapper = ['Demographic_key.csv', 'Diet_key.csv', 'Examination_key.csv', 'Lab_key.csv', 'Questionnaire_key.csv']
```

```python
#Function
def to_dictionary(df):
    if len(df.columns) == 1:
        if df.values.size == 1: 
            return df.values[0][0]
        return df.values.squeeze()
    grouped = df.groupby(df.columns[0])
    to_dict = {k: recur_dictify(g.ix[:,1:]) for k,g in grouped}
    return to_dict
```

```python
for dataset in cleaned_datasets:
    to_be_mapped_df = pd.read_csv(dataset)
    export_name = "mapped_" + dataset[:-13]
    index = cleaned_datasets.index(dataset)
    mapper_df = pd.read_csv(mapper[index])
    mapper_df = mapper_df.loc[:, ['short_title','Lookup', 'Column']]
    to_dict = to_dictionary(mapper_df)
    range_list = []
    for k,v in to_dict.items():
        for k,v in v.items():
            if('Range of Values' in v):
                range_list.append(k.split('_')[0])
    for col in to_be_mapped_df.columns[1:]:
        try:
            if(col not in range_list):
                to_be_mapped_df[col] = to_be_mapped_df[col].map(to_dict[col])
            else:
                to_be_mapped_df[col] = to_be_mapped_df[col].str.split('_').str[1]
        except:
            print(col)
    to_be_mapped_df.to_csv(export_name + ".csv", index=False)
    print(export_name)

```
